```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

# Simple Linear Regression

```
drive_file = "drive/My Drive/Colab Notebooks/Datasets/student_scores.csv"
st_score = pd.read_csv(drive_file)
```

```
st_score
x = st_score.iloc[:, :-1]
y = st_score.iloc[:, 1]
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```
l_reg = LinearRegression().fit(x_train, y_train)
```

```
predicted_values = l_reg.predict(x_test)

prediction_df = pd.DataFrame({
    "Real Value": y_test,
    "Predicted Value": predicted_values
})
```

```
prediction_df

prediction = l_reg.predict(x)
```

```
plt.plot(x, y, "o")
plt.title("Studant grade per hour studied")
plt.xlabel("Hours Studied")
plt.ylabel("Student Grade")
plt.plot(x, prediction)
```

# Multiple Linear regression

```
petrol_cons = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/petrol_consumption.csv")
```

```
X = petrol_cons[["Petrol_tax", "Average_income", "Paved_Highways",
       "Population_Driver_licence(%)"]]
y = petrol_cons["Petrol_Consumption"]
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```
mult_model = LinearRegression().fit(X_train, y_train)

predicted_value = mult_model.predict(X_test)
```

```
prediction_comp = pd.DataFrame({
    "Real Value": y_test,
    "Predicted_value": predicted_value
})

prediction_comp
```
